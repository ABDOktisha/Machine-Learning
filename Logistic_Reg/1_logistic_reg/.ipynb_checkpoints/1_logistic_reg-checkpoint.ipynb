```python
def drop_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

df = drop_outliers(df, 'time_spend_company')
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['time_spend_company'])
plt.title("count of years in the companey", fontsize=14)
plt.xlabel("years", fontsize=12)
plt.grid(axis='x', alpha=0.5)
plt.show()
```

```python
def count_category(df, column):
    counts = df[column].value_counts().sort_index()

    counts_sorted = counts.sort_values(ascending=True)

    colors = sns.color_palette("Blues", len(counts_sorted))

    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, data=df, order=counts_sorted.index, palette=dict(zip(counts_sorted.index, colors)))  # Use the column variable here
    plt.title(f'Distribution of {column}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
count_category(df, 'department')
```

```python
count_category(df, 'salary')
```
